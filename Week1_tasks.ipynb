```python
# Your code starts here
titles = df_train['Title'].unique()
num_sur = []
num_die = []
num_total = []
survival_rate = []
for title in titles:
  sur_ctr = 0
  die_ctr = 0
  for i in range(df_train.shape[0]):
    if(df_train['Title'][i] == title):
      if(df_train['Survived'][i]==1):
        sur_ctr+=1
      elif(df_train['Survived'][i]==0):
        die_ctr+=1
  num_sur.append(sur_ctr)
  num_die.append(die_ctr)
num_total = list(map(sum,zip(num_sur,num_die)))
for i in range(len(num_sur)):
  survival_rate.append(num_sur[i]/num_total[i])

dataList = zip(titles,survival_rate)
SR_df = pd.DataFrame(dataList,columns = ['Title','Survival Rate'])

# Your code ends here
```

#Standardization vs Normalization

#Feature Scaling

In machine learning, most of the supervised and unsupervised learning methods rely on data sets provided in the training model, and desicions made by algorithms is done by measuring the Euclidean distance between data points in order to better understand the data. Sometimes we have data sets which have large variance and magnitude and could end up giving a larger influence to the training of the model than is expected. Thus if we find ways to bring the values of the features close to each other, there are chances for the algorithm to get trained well and faster instead of the data set where the data points or features values have high differences with each other will take more time to understand the data and the accuracy will be lower. 

This brings us to scaling, which is essentially a technique to bring data points closer together or in other words, reduce the distance between the data points.Normalization and Standardization are the two main methods for the scaling of the data. Let us compare both of them.

#Normalization 

Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1. 

Geometrically speaking, transformation squishes the n-dimensional data into   an n-dimensional unit hypercube. The new point is calculated as:

``X_new = (X - X_min)/(X_max - X_min)``       

Normalization is a useful tool when there are no outliers in the data.

It is generally utilised for data sets which do not follow a Gaussian distribution.

#Standardization 
Standardization is another scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.

Geometrically speaking, it translates the data to the mean vector of original data to the origin and squishes or expands the points if std is 1 respectively. It is given by:

``X_new = (X - mean)/Std``

Standardization is not much affected by outliers since it does not produce a predefined range of the data altered.

Standardization can be helpful in cases where the data follows a Gaussian Distribution.
